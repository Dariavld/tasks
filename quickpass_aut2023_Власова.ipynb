#### Задача 1 (6 баллов). 

Напишем игру "ферма". Наш игрок - экземпляр класса "фермер", у которого есть имя, стартовая сумма денег и объект класса "ферма". У объекта класса "ферма" есть жилплощадь (для самого фермера), а также могут быть поле для картошки, например (с площадью), огород под огурцы, постройки для домашних животных (с вместимостью). Наш фермер может покупать семена и животных, собирать урожай и взаимодействовать с животными (коров можно доить, свиней можно зарезать и продать мясо). Игра консольная, поэтому придется выводить нужную информацию через запросы игрока (игрок может захотеть знать, сколько у него есть животных, какие у него есть огороды-поля и так далее). 

Можно не строго следовать описанию, а использовать фантазию. Если мне очень понравится, могу приплюсовать балл.

```python
class Farmer:
    def __init__(self, name, money):
        self.name = name
        self.money = money
        self.farm = Farm(self)

class Farm:
    def __init__(self, farmer):
        self.farmer = farmer
        self.living_space = 1
        self.potato_field = 0
        self.cucumber_garden = 0
        self.animal_buildings = {'корова': 0, 'свинья': 0}
        self.animals = {'корова': 0, 'свинья': 0}

    def buy_seed(self, seed_type, quantity, price):
        if self.farmer.money >= price:
            self.farmer.money -= price
            if seed_type == 'картошка':
                self.potato_field += quantity
            elif seed_type == 'огурец':
                self.cucumber_garden += quantity
            print(f"Купил {quantity} семена следующего вида - {seed_type} за {price} рублей.")
        else:
            print("Недостаточно денег, чтобы купить.")

    def buy_animal(self, animal_type, quantity, price):
        if self.farmer.money >= price:
            self.farmer.money -= price
            self.animal_buildings[animal_type] += quantity
            self.animals[animal_type] += quantity
            print(f"Купил {quantity} животину следующего вида - {animal_type} за {price} dollars.")
        else:
            print("Недостаточно денег, чтобы купить.")

    def harvest_crop(self, crop_type):
        if crop_type == 'картошка':
            if self.potato_field > 0:
                self.potato_field -= 1
                self.farmer.money += 10
                print("Собрал урожай картофеля.")
            else:
                print("Еще нет урожая.")
        elif crop_type == 'огурец':
            if self.cucumber_garden > 0:
                self.cucumber_garden -= 1
                self.farmer.money += 15
                print("Собрал урожай огурцов.")
            else:
                print("Еще не время собирать урожай.")

    def interact_animal(self, animal_type):
        if self.animals[animal_type] > 0:
            if animal_type == 'корова':
                self.farmer.money += 20
                print("Подоил корову.")
            elif animal_type == 'свинья':
                self.farmer.money += 50
                print("Зарезал свинью, продал мясо.")
            self.animals[animal_type] -= 1
        else:
            print(f"Нет животины - {animal_type} для взаимодействия.")

    def display_info(self):
        print(f"Жилплощадь: {self.living_space}")
        print(f"Грядка с картошкой: {self.potato_field}")
        print(f"Огород с огурцами: {self.cucumber_garden}")
        print(f"Загоны для коровы: {self.animal_buildings['корова']}")
        print(f"Свинарники: {self.animal_buildings['свинья']}")
        print(f"Коровы: {self.animals['корова']}")
        print(f"Свиньи: {self.animals['свинья']}")
```

Напишите программу - примитивного голосового помощника, задача которого - помогать пользователю изучать выбранный язык. Такая программа должна уметь следующие вещи: 

- разговаривать с пользователем голосом (и распознавать его речь)
- знать команды "переведи" и "зачитай вслух". Обе команды заставляют программу предложить пользователю ввести текст, который потом переводится и/или зачитывается программой. 
- знать команду "практика": программа предлагает пользователю небольшой текст, выбранный случайным образом из базы данных программы. Пользователь должен зачитать текст вслух, а программа должна сравнить распознанный голос пользователя с исходным текстом и сообщить, если не совпало. 
- опционально можно добавить функцию для ручного пополнения базы данных. 

Язык выбираете какой угодно (из поддерживаемых библиотеками, конечно). 

```python
import speech_recognition as sr
import pyttsx3
import googletrans
import random


recognizer = sr.Recognizer()
engine = pyttsx3.init()
translator = googletrans.Translator()

#распознавание речи
def recognize_speech():
    with sr.Microphone() as source:
        print("Говорите..")
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            print("Вы сообщили: " + text)
            return text
        except sr.UnknownValueError:
            print("Извините, не распознал ваши слова.")
        except sr.RequestError as e:
            print("Ошибка с Гуглом; {0}".format(e))

#прочитать текст вслух
def read_aloud(text):
    engine.say(text)
    engine.runAndWait()

#перевод текста
def translate_text():
    print("Что хотите перевести?")
    text = recognize_speech()
    print("На какой язык?")
    language = recognize_speech()
    translation = translator.translate(text, dest=language).text
    read_aloud(translation)

#тренировка речи
def practice_speaking():
    texts = ["Quick Brown Fox Jumps Over The Lazy Dog.", "Who wrote the famous sentence about quick foxes and lazy dogs?", "The quick brown fox alternatives"]
    text = random.choice(texts)
    print("Пожалуйста, прочитайте вслух:")
    print(text)
    read_aloud(text)
    print("Правильно прочитали?")
    answer = recognize_speech()
    if answer.lower() == "да":
        print("Отлично, поздравляю!")
    else:
        print("Попробуйте еще раз.")

#добавление текста в бд
def add_text():
    print("Введите текст, который хотите добавить в базу данных:")
    text = input()
    

#основной сценарий общения с пользователем 
while True:
    print("Чем могу помочь?")
    command = recognize_speech()
    if "переведи" in command.lower():
        translate_text()
    elif "зачитай вслух" in command.lower():
        print("Что вы хотите, чтобы я зачитал вслух?")
        text = recognize_speech()
        read_aloud(text)
    elif "практика" in command.lower():
        practice_speaking()
    elif "добавь текст" in command.lower():
        add_text()
    elif "конец" in command.lower():
        break
    else:
        print("Извините, не распознал вашу команду.")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import ipywidgets as widgets
from IPython.display import display

def plot_word_frequency(text, n):
    words = text.split()
    word_counts = Counter(words)
    top_words = word_counts.most_common(n)
    
    #график
    fig, ax = plt.subplots()
    ax.bar([word[0] for word in top_words], [word[1] for word in top_words])
    ax.set_xlabel("Слово")
    ax.set_ylabel("Частота")
    ax.set_title("{} самых частотных слов".format(n))
    plt.show()


text = '''Портал Проза.ру предоставляет авторам возможность свободной публикации своих литературных произведений в сети Интернет на основании пользовательского договора. 
Все авторские права на произведения принадлежат авторам и охраняются законом. 
Перепечатка произведений возможна только с согласия его автора, к которому вы можете обратиться на его авторской странице. 
Ответственность за тексты произведений авторы несут самостоятельно на основании правил публикации и законодательства Российской Федерации. Данные пользователей обрабатываются на основании Политики обработки персональных данных. Вы также можете посмотреть более подробную информацию о портале и связаться с администрацией.
Ежедневная аудитория портала Проза.ру – порядка 100 тысяч посетителей, 
которые в общей сумме просматривают более полумиллиона страниц по данным счетчика посещаемости, который расположен справа от этого текста. 
В каждой графе указано по две цифры: количество просмотров и количество посетителей. Все права принадлежат авторам, 2000-2023.   
Портал работает под эгидой Российского союза писателей.'''

def update_plot(n):
    plot_word_frequency(text, n)

slider = widgets.IntSlider(min=1, max=10, value=5)
widgets.interact(update_plot, n=slider)
```

#### Задача 5 (1 балл).

Напишите программу, которая будет запрашивать у пользователя, какой из трех любых (выберите сами) признаков WALS он бы хотел посмотреть, и отрисовывать карту с WALS с этим признаком при помощи folium. Маркеры разных значений признака должны иметь разный цвет и иконку. 

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from ipywidgets import interact, Dropdown


wals_df = pd.read_csv('https://github.com/cldf-datasets/wals/blob/master/cldf/languages.csv')


def create_wals_map(feature):
    marker_cluster = MarkerCluster()

   
    for index, row in wals_df.iterrows():
        value = row[feature]

         #разные цвета и иконки
        if value == 0:
            color = 'gray'
            icon = 'question'
        elif value == 1:
            color = 'blue'
            icon = 'info-sign'
        elif value == 2:
            color = 'green'
            icon = 'ok-sign'
        else:
            color = 'red'
            icon = 'exclamation-sign'

        
        marker = folium.Marker(location=[row['Latitude'], row['Longitude']],
                               icon=folium.Icon(color=color, icon=icon))


        marker_cluster.add_child(marker)

    
    map = folium.Map(location=[wals_df.iloc[0]['Latitude'], wals_df.iloc[0]['Longitude']], zoom_start=2)
    map.add_child(marker_cluster)
    return map




features = ['81A Order of Subject, Object and Verb', '81B Order of Subject, Verb and Object', '81C Order of Object and Verb']
dropdown = Dropdown(options=features)

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        feature = change['new']
        map = create_wals_map(feature)
        display(map)

dropdown.observe(on_change)
display(dropdown)
```

Задание 16. Переведем баллы в оценки

Система перевода баллов в оценки

      [90;100] = A
      [80;90) = B
      [70;80) = C
      [60;70) = D
      [50;60) = E
      [0;50) = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return "A"
    elif average_mark >= 80:
        return "B"
    elif average_mark >= 70:
        return "C"
    elif average_mark >= 60:
        return "D"
    elif average_mark >= 50:
        return "E"
    else:
        return "F"

df1["Grade"] = df1.apply(lambda x: GetGrade((x["math score"] + x["reading score"] + x["writing score"]) / 3), axis=1)


start_time = time.time()
grade_counts = df1["Grade"].value_counts()

print("Количество студентов, получивших каждую из оценок:")
print(grade_counts)
print("Время исполнения:", time.time() - start_time, "секунд")
```

#### Задача 8 (5 баллов). 

Некоторое лингвистическое исследование показало, что в дневниковых записях и подобных текстах (н-р, блоги ЖЖ) на 100 личных местоимений местоимения распределяются следующим образом:

    я - 50 раз
    ты - 5 раз
    он/она/оно - 25 раз
    мы - 3 раза
    вы - 2 раза
    они - 15 раз
    
Возьмите любой достаточно большой набор текстов (блоги ЖЖ, дневниковые записи) и проверьте, действительно ли в нем такое распределение местоимений. 

```python
import os
import re

directory = 'D:\Data'
filenames = ['LJ1.txt', 'LJ2.txt', 'LJ3.txt']

texts = []
for filename in filenames:
    filepath = os.path.join(directory, filename)
    with open(filepath, 'r', encoding='utf-8') as file:
        text = file.read()
        texts.append(text)

pronoun_counts = {
    'я': 0,
    'ты': 0,
    'он/она/оно': 0,
    'мы': 0,
    'они': 0
}

for text in texts:
    words = re.findall(r'\b\w+\b', text)
    for word in words:
        if word.lower() in pronoun_counts:
            pronoun_counts[word.lower()] += 1


total_pronouns = sum(pronoun_counts.values())
if total_pronouns == 0:
    pronoun_distribution = {k: 0 for k in pronoun_counts.keys()}
else:
    pronoun_distribution = {k: v / total_pronouns * 100 for k, v in pronoun_counts.items()}


expected_distribution = {
    'я': 50,
    'ты': 5,
    'он/она/оно': 25,
    'мы': 3,
    'они': 15
}

for pronoun, expected_count in expected_distribution.items():
    calculated_count = pronoun_distribution[pronoun]
    print(f"{pronoun}: expected {expected_count}%, calculated {calculated_count:.2f}%")
```
